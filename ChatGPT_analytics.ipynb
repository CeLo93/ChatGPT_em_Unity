```python
import openai
import matplotlib.pyplot as plt

# Configure sua chave de API
openai.api_key = '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'

# Função para interagir com o ChatGPT
def interagir_com_chat(prompt):
    resposta = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",  # Escolha o mecanismo adequado
        messages=[{"role": "system", "content": "Você é um ChatGPT"}, {"role": "user", "content": prompt}],
    )
    return resposta.choices[0].message['content']

# Prompt da conversa
prompt_conversa = "Qual é a reputação da PUC Minas?"

# Interaja com o ChatGPT
resposta_do_chat = interagir_com_chat(prompt_conversa)

# Imprima a resposta
print("Resposta do ChatGPT:", resposta_do_chat)

# Análise de Comprimento de Resposta
comprimento_resposta = len(resposta_do_chat.split())

# Histograma de Padrões de Resposta
padroes_resposta = resposta_do_chat.split('.')
num_padroes = len(padroes_resposta)

# Análise de Sentimento (simulação)
sentimento = 0.8  # Valor simulado de sentimento (positivo)

# Plotagem das análises
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Análise da Conversa')

# Histograma de Comprimento de Resposta
axs[0, 0].bar("Resposta", comprimento_resposta, color='blue')
axs[0, 0].set_title('Análise de Comprimento de Resposta')
axs[0, 0].set_xlabel('Resposta')
axs[0, 0].set_ylabel('Número de Palavras')

# Histograma de Padrões de Resposta
axs[0, 1].bar("Padrões de Resposta", num_padroes, color='green')
axs[0, 1].set_title('Histograma de Padrões de Resposta')
axs[0, 1].set_xlabel('Número de Padrões')
axs[0, 1].set_ylabel('Frequência')

# Análise de Sentimento
axs[1, 0].bar("Sentimento", sentimento, color='orange')
axs[1, 0].set_title('Análise de Sentimento')
axs[1, 0].set_xlabel('Sentimento')
axs[1, 0].set_ylabel('Valor')

# Exibição dos gráficos
plt.tight_layout()
plt.show()
```
